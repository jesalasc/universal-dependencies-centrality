# Análisis de frecuencia de verbos en oraciones simples

```python
import pandas as pd
df_class = pd.read_csv("clasificacion_resultados3.csv")
print(df_class.shape[0])
df_class["ID"] -= 1
string = "es_gsd-ud-train_{}.graphml"
df_class["filename"] = df_class["ID"].apply(lambda x: string.format(x))
df_class = df_class[df_class["Clasificación"] == "Simple"]
df_class
```

```python
import networkx as nx
def get_verb(row):
    try:
        graph = nx.read_graphml(f"../data/graphs_labeled_filtered/UD_Spanish-GSD/{row['filename']}")
    except Exception as e:
        return None
    labels = nx.get_node_attributes(graph, 'label')  # node labels
    forms = nx.get_node_attributes(graph, 'lemma')  # node labels
    #colors = [labels[u] for u in G.nodes()] 
    verbs = [forms[n] for n in labels if labels[n] in ["VERB", "AUX"]]
    return verbs[0]

df_class["verb"] = df_class.apply(get_verb, axis=1)
df_class = df_class.dropna(subset=["verb"])
df_class
```

```python
df_class.groupby("verb").size().sort_values(ascending=False).reset_index().rename(columns={0:"frequency"}).to_csv("frequency_verbs_simple_sentences.csv", sep=";", index=False)
```
